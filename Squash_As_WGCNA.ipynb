```python
import PyWGCNA as wgcna
import pandas as pd
```

```python
pepo = "C:Desktop/GerardoIga/2024/CucurbitapepoAs/CpepoRNAseq/C_pepo_TPMs_As.csv"
pepo = pd.read_csv(pepo)
pepo.head()
pepo = pepo.transpose()
pepo.columns = pepo.iloc[0]
pepo = pepo[1:]
pepo.to_csv("pepoTransposed.csv")
pepo = "pepoTransposed.csv"
pepo
```

```python
pepoWGCNA = wgcna.WGCNA(name='pepo', 
                              species='Cucurbita pepo', 
                              geneExpPath=pepo, 
                              outputPath='',
                              save=True,
                                )
```
